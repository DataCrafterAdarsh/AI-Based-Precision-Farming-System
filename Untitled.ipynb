```python
import pandas as pd
import random
import time
from datetime import datetime

file_name = "sensor_data.csv"

# Create CSV if not exists
try:
    pd.read_csv(file_name)
except FileNotFoundError:
    df = pd.DataFrame(columns=[
        "Time", "Temperature_C", "Humidity_%", "Soil_Moisture_%"
    ])
    df.to_csv(file_name, index=False)

print("Simulating sensor data... Press Stop to end")

for _ in range(20):   # generates 20 readings
    temperature = round(random.uniform(26, 35), 1)
    humidity = random.randint(40, 75)
    soil = random.randint(25, 60)

    new_row = {
        "Time": datetime.now(),
        "Temperature_C": temperature,
        "Humidity_%": humidity,
        "Soil_Moisture_%": soil
    }

    df = pd.read_csv(file_name)
    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
    df.to_csv(file_name, index=False)

    new_row
    time.sleep(2)
```
